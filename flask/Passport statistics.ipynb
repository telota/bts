```python
import sqlite3
import collections
import json
from IPython import display
```

```python
db = sqlite3.connect('data.sqlite3')
```

```python
meta = [ j for j in (json.loads(data) for data, in db.execute('SELECT metadata FROM corpus_object')) if j ] 
```

```python
display.display(meta[10002])
```

```python
def paths(obj):
    if not isinstance(obj, dict):
        return
    
    for k, vs in obj.items():
        yield k
        for v in vs:
            for path in paths(v):
                yield f'{k}/{path}'
```

```python
collections.Counter(path for obj in meta for path in paths(obj))
```

```python
def path_counts(obj):
    if not isinstance(obj, dict):
        return
    
    for k, vs in obj.items():
        yield len(vs), f'{k}'
        for v in vs:
            for n, path in path_counts(v):
                yield n, f'{k}/{path}'
```

```python
sorted(collections.Counter(path for obj in meta for path in path_counts(obj)))
```
